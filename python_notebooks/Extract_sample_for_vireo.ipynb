```python
def ret_samp_names(y, df_info):
    #print(len(y), len(bc))
    if df_info.loc[df_info['SubID'] == y, 'cDNA_ID'].isnull().all():
        return ""
    else:
        return '-'.join(df_info.loc[df_info['SubID'] == y, 'cDNA_ID'].to_list()[0].split('-')[:3])[:-1]
```
